```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


data = pd.read_csv('Airplane_Crashes_and_Fatalities_Since_1908.csv')
data.info()
```

## Bagi dataset menjadi training set dan testing set dengan proporsi 70:30

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns=['Aboard'])
y = data['Aboard']
# Pisahkan dataset menjadi training set (70%) dan testing set (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

data.head()
```

# Lakukan normalisasi data pada salah satu attribute menggunakan Min Max scaler (buatlah copy dataset terlebih dahulu)

```python
copydata = data.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler

kolom_normalisasi = ['Aboard']


scaler = MinMaxScaler()

copydata[kolom_normalisasi] = scaler.fit_transform(copydata[kolom_normalisasi])
copydata
```

# Lakukan standarisasi pada dataset

```python
from sklearn.preprocessing import StandardScaler
copydata['Aboard'] = pd.to_numeric(data['Aboard'], errors='coerce', downcast='integer')
kolom_standarisasi = [ 'Aboard', 'Aboard Passangers', 'Aboard Crew', 'Fatalities']
scaler = StandardScaler()
copydata[kolom_standarisasi] = scaler.fit_transform(copydata[kolom_standarisasi])
copydata
```

# Lakukan Data cleaning pada data dengan nilai null

```python
# Misalnya, kita mengganti nilai null dengan median, mean, atau modus sesuai dengan jenis atributnya
copydata.fillna({'numeric_column': copydata['Aboard'].median(),
                     'float_column': copydata['Fatalities'].mean(),
                     'categorical_column': copydata['Ground'].mode()[0]}, inplace=True)
```

```python
copydata.isna().sum()
```

# Lakukan Data cleaning pada data dengan nilai duplikat. (Jika tidak ada nilai duplikat pada dataset, maka buatlah menjadi ada)

```python
duplicatedata = data[data.duplicated(keep=False)]

duplicatedata
```

```python
banyak_duplikat = 5
duplicate = pd.concat([copydata] * banyak_duplikat, ignore_index=True)
```

```python
data_duplikat = duplicate[duplicate.duplicated(keep=False)]
data_duplikat
```

```python
cleandata = data_duplikat.drop_duplicates()
cleandata
```

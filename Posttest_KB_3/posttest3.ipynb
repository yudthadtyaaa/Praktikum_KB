# Grafik Batang

```python

x = list(new_df.iloc[:,8])
y = list(new_df.iloc[:,1])
```

```python
# plt.figure(figsize=(5,6))
plt.figure(figsize=(11,11))
plt.bar(x, y)
plt.title("Google Play App Rating")
plt.xlabel("Content Rating")
plt.ylabel("Category")
```

### Gambar diatas menampilkan berapa banyak Content Rating dari masing - masing kategori Aplikasi tersebut

```python
df = df.dropna(subset=['Type'])
```

# Grafik Pie Chart

```python
pie = df['Type'].value_counts().rename_axis('Type').reset_index(name='Rating')
```

```python
plt.figure(figsize=(10,10))
plt.pie(pie.Rating, labels=pie.Type, startangle=90,autopct='%.1f%%')
plt.title("Rating")
plt.show()
```

### Dari gambar diatas banyak sekali Aplikasi yang Free dibandingkan dengan berbayar

# Grafik Scatter Plot

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data = df, x = "Type", y = "Rating")
plt.show()
```

### Sama seperti Pie Chart, sedikit lebih banyak aplikasi yang free dibandingkan berbayar

# Grafik Line Pot

```python
plt.figure(figsize=(11,11))
plt.plot(df.Type, df.Category)
```

### Diatas diagram line yang sama memiliki Tipe free, paid sama nol nah disini banyak sekali interaksi antara free dan paid sementara yang nol hanya satu saja

# Grafik Heatmap

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), center=0, cmap='Blues')
ax.set_title('Korelasi antar atribut pada googleplaystore.csv')
```

### Data saya tidak bisa menggunakan Heatmap

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("WineQT.csv")
```

```python
# Variabel (X) dan target (y)
X_reg = df[['sulphates', 'alcohol']]
y_reg = df['quality']
```

```python
# Bagi dataset menjadi data latih dan data uji
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
```

```python
# Inisialisasi model Linear Regression
linear_reg = LinearRegression()
linear_reg.fit(X_train_reg, y_train_reg)
```

```python
# Prediksi kualitas anggur
y_pred_reg = linear_reg.predict(X_test_reg)
```

```python
# Evaluasi Metode Regresi
# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test_reg, y_pred_reg)
print("Mean Absolute Error:", mae)

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test_reg, y_pred_reg)
print("Mean Squared Error:", mse)

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)

# R-squared (R2) Score
r2 = r2_score(y_test_reg, y_pred_reg)
print("R-squared (R2) Score:", r2)
```

```python
# A3.2: Plot hasil prediksi
plt.figure(figsize=(12, 6))
plt.scatter(y_test_reg, y_pred_reg)
plt.xlabel("Nilai Qualitas (Asli)")
plt.ylabel("Nilai Qualitas (Prediksi)")
plt.title("Linear Regression - Nilai Qualitas (Asli) vs Nilai Qualitas (Prediksi)")
plt.show()
```
